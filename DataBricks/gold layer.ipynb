```python
silver_path = "/Volumes/nyc/default/silver/"
gold_path = "dbfs:/Volumes/nyc/default/gold"
```

```python
%sql
CREATE SCHEMA nyc.gold;
```

```python
%sql
USE nyc.gold;
```

```python
from pyspark.sql.functions import *
```

```python
df_zone = spark.read.format('parquet').load(silver_path + 'trip_zone')
```

```python
df_zone.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("nyc.gold.trip_zone")
```

```python
gold_path = "abfss://gold@saazurenycproject.dfs.core.windows.net/trip_zone"

df_zone.write \
    .format("delta") \
    .mode("overwrite") \
    .save(gold_path)
```

```python
df_trip_type = spark.read.format('parquet').load(silver_path + 'trip_type')
```

```python
df_trip_type.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("nyc.gold.trip_type")
```

```python
gold_path = "abfss://gold@saazurenycproject.dfs.core.windows.net/trip_type"

df_trip_type.write \
    .format("delta") \
    .mode("overwrite") \
    .save(gold_path)
```

```python
df_trips2023data = spark.read.format('parquet').load(silver_path + 'trips2023data')
```

```python
df_trips2023data.write \
    .format("delta") \
    .mode("overwrite") \
    .option("mergeSchema", "true")\
    .saveAsTable("nyc.gold.trips2023data")
```
